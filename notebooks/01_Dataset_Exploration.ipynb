```python
zip_files = sorted(ARCHIVE_FOLDER.glob("*.zip"))

print(f"Found {len(zip_files)} weekly zip files.\n")

for z in zip_files:
    print(z.name)
```

```python
for zip_path in zip_files:

    extract_path = OUTPUT_FOLDER / zip_path.stem

    if extract_path.exists():
        print(f"Already extracted: {zip_path.name}")
        continue

    print(f"Extracting {zip_path.name}")

    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(extract_path)

print("\nAll ZIP files extracted successfully.")
```

```python
excel_files = []

for file in OUTPUT_FOLDER.rglob("*"):

    if file.suffix.lower() in [".xlsx", ".xls"]:

        excel_files.append(file)

print(f"Total Excel Files: {len(excel_files)}")
```
