```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

plt.style.use("ggplot")

pd.set_option("display.max_columns",None)
```

```python
PROJECT_DIR = Path.cwd().parent

DATA = PROJECT_DIR/"processed"/"cleaned_dataset.csv"

df = pd.read_csv(DATA)

print(df.shape)

df.head()
```

```python
print("="*60)

print("Dataset Shape :",df.shape)

print("="*60)

df.info()
```

```python
df.describe(include="all")
```

```python
missing = pd.DataFrame({

    "Missing":df.isnull().sum(),

    "Percentage":round(df.isnull().mean()*100,2)

})

missing.sort_values(

    "Missing",

    ascending=False

)
```

```python
plt.figure(figsize=(10,5))

df["vehicle_category"].value_counts().plot(

    kind="bar"

)

plt.title("Vehicle Category Distribution")

plt.xlabel("Vehicle Category")

plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

df["measurement_site"].value_counts().plot(

    kind="bar"

)

plt.title("Measurement Site Distribution")

plt.ylabel("Records")

plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(

    data=df,

    x="peak_period"

)

plt.title("Peak Period Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(

    df["total_traffic"],

    bins=50,

    kde=True

)

plt.title("Distribution of Total Traffic")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(

    data=df,

    x="vehicle_category",

    y="total_traffic"

)

plt.xticks(rotation=45)

plt.title("Traffic Distribution by Vehicle Category")

plt.show()
```

```python
site = (

    df.groupby("measurement_site")["total_traffic"]

    .sum()

    .sort_values(ascending=False)

    .head(10)

)

plt.figure(figsize=(12,6))

site.plot(kind="bar")

plt.ylabel("Total Traffic")

plt.title("Top 10 Busiest Measurement Sites")

plt.show()
```

```python
hourly = (

    df.groupby("hour")["total_traffic"]

    .mean()

)

plt.figure(figsize=(10,5))

plt.plot(

    hourly.index,

    hourly.values,

    marker="o"

)

plt.title("Average Hourly Traffic")

plt.xlabel("Hour")

plt.ylabel("Average Traffic")

plt.grid(True)

plt.show()
```

```python
traffic_cols = [

    c

    for c in df.columns

    if c.startswith("direction_")

]

corr = df[traffic_cols+["total_traffic"]].corr()

plt.figure(figsize=(12,10))

sns.heatmap(

    corr,

    cmap="coolwarm",

    center=0

)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(

    y=df["total_traffic"]

)

plt.title("Outlier Detection")

plt.show()
```

```python
peak = (

    df.groupby("peak_period")["total_traffic"]

    .mean()

)

peak
```

```python
pivot = pd.pivot_table(

    df,

    values="total_traffic",

    index="vehicle_category",

    columns="peak_period",

    aggfunc="mean"

)

pivot
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(

    pivot,

    annot=True,

    cmap="YlGnBu"

)

plt.title("Average Traffic")

plt.show()
```
