```python
import pandas as pd
import numpy as np
from pathlib import Path
import os

pd.set_option("display.max_columns", None)
```

```python
PROJECT_DIR = Path.cwd().parent

DATA_FOLDER = PROJECT_DIR / "Extracted_Data"

OUTPUT_FOLDER = PROJECT_DIR / "processed"

OUTPUT_FOLDER.mkdir(exist_ok=True)
```

```python
excel_files = sorted(DATA_FOLDER.rglob("*.xlsx"))

print(f"Found {len(excel_files)} Excel files")
```

```python
#we skip document sheets
IGNORE_SHEETS = [
    "Erklärungen",
    "Explanation",
    "Info"
]
```

```python
master_df = []

for file in excel_files:

    print(f"Reading: {file.name}")

    try:
        xls = pd.ExcelFile(file)

    except Exception as e:
        print(e)
        continue

    week = file.parents[1].name
    site = file.parent.name

    for sheet in xls.sheet_names:

        if sheet in IGNORE_SHEETS:
            continue

        try:

            df = pd.read_excel(
                file,
                sheet_name=sheet,
                header=12
            )

            # Remove empty rows
            df.dropna(how="all", inplace=True)

            # Remove empty columns
            df.dropna(axis=1, how="all", inplace=True)

            # Add metadata
            df["Vehicle_Category"] = sheet
            df["Week"] = week
            df["Measurement_Site"] = site
            df["Source_File"] = file.name

            master_df.append(df)

        except Exception as e:

            print(f"Skipped {sheet}")

            print(e)
```

```python
master_df = pd.concat(master_df, ignore_index=True)

print(master_df.shape)

master_df.head()
```

```python
#rename columns for better understanding
master_df.rename(columns={

    "Startzeit":"Time",

    "Intervall":"Total",

    "Summe Stunde":"Hourly_Total"

}, inplace=True)
```

```python
master_df.columns = (
    master_df.columns
    .astype(str)
    .str.strip()
    .str.replace("\n"," ")
)
```

```python
def split_location(site):

    try:

        tokens = site.rsplit(" ",1)

        location = tokens[0]

        coords = tokens[1]

        lat, lon = coords.split(",")

        return pd.Series([location,float(lat),float(lon)])

    except:

        return pd.Series([site,np.nan,np.nan])


master_df[
    ["Location","Latitude","Longitude"]
] = master_df["Measurement_Site"].apply(split_location)
```
