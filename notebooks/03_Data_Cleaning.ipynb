```python
import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option("display.max_columns", None)
```

```python
PROJECT_DIR = Path.cwd().parent

DATA_PATH = PROJECT_DIR / "processed" / "master_dataset.csv"

df = pd.read_csv(DATA_PATH)

print(df.shape)

df.head()
```

```python
print("="*60)

print(df.info())

print("="*60)

print(df.describe(include="all"))
```

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": round(df.isnull().mean()*100,2)
})

missing.sort_values(
    "Missing Values",
    ascending=False
).head(20)
```

```python
#remove duplicate rows
print("Before :", len(df))

df = df.drop_duplicates()

print("After :", len(df))
```

```python
df.columns = (

    df.columns

    .str.strip()

    .str.replace(" ", "_")

    .str.replace("-", "_")

    .str.replace("/", "_")

    .str.replace("(", "")

    .str.replace(")", "")

    .str.lower()
)

df.columns.tolist()
```

```python
meta_cols = [

    "week",

    "measurement_site",

    "latitude",

    "longitude",

    "vehicle_category",

    "source_file",

    "time"
]
```

```python
numeric_cols = [

    c

    for c in df.columns

    if c not in meta_cols
]
```

```python
for col in numeric_cols:

    df[col] = pd.to_numeric(

        df[col],

        errors="coerce"
    )
```

```python
df.dtypes
```

```python
df["time"] = pd.to_datetime(
    df["time"],
    format="%H:%M:%S",
    errors="coerce"
)
```

```python
df["hour"] = df["time"].dt.hour

df["minute"] = df["time"].dt.minute
```

```python
def peak_period(hour):

    if 6 <= hour < 10:
        return "Morning Peak"

    elif 10 <= hour < 15:
        return "Midday"

    elif 15 <= hour < 19:
        return "Evening Peak"

    else:
        return "Off Peak"
```

```python
df["peak_period"] = df["hour"].apply(peak_period)
```

```python
traffic_cols = [

    c

    for c in df.columns

    if c.startswith("direction_")
]
```

```python
df["total_traffic"] = (

    df[traffic_cols]

    .sum(axis=1)
)
```

```python
OUTPUT = PROJECT_DIR / "processed"

OUTPUT.mkdir(exist_ok=True)

df.to_csv(

    OUTPUT / "cleaned_dataset.csv",

    index=False
)

print("Saved Successfully!")
```
