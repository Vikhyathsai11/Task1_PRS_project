```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

plt.style.use("ggplot")

pd.set_option("display.max_columns",None)
```

```python
PROJECT_DIR = Path.cwd().parent

DATA = PROJECT_DIR/"processed"/"cleaned_dataset.csv"

df = pd.read_csv(DATA)

print(df.shape)

df.head()
```

```python
print(df["vehicle_category"].unique())
```

```python
hourly_profile = (

    df.groupby(

        ["vehicle_category","hour"]

    )["total_traffic"]

    .mean()

    .reset_index()

)

hourly_profile.head()
```

```python
categories = hourly_profile["vehicle_category"].unique()

for category in categories:

    plt.figure(figsize=(10,5))

    temp = hourly_profile[
        hourly_profile["vehicle_category"]==category
    ]

    plt.plot(

        temp["hour"],

        temp["total_traffic"],

        marker="o"

    )

    plt.title(f"Daily Traffic Profile - {category}")

    plt.xlabel("Hour")

    plt.ylabel("Average Traffic")

    plt.xticks(range(24))

    plt.grid(True)

    plt.show()
```

```python
peak_hours = (

    hourly_profile

    .loc[

        hourly_profile

        .groupby("vehicle_category")

        ["total_traffic"]

        .idxmax()

    ]

)

peak_hours
```

```python
plt.figure(figsize=(10,5))

sns.barplot(

    data=peak_hours,

    x="vehicle_category",

    y="total_traffic"

)

plt.xticks(rotation=45)

plt.title("Peak Traffic by Vehicle Category")

plt.ylabel("Average Traffic")

plt.show()
```

```python
def traffic_period(hour):

    if 6<=hour<10:

        return "Morning"

    elif 15<=hour<19:

        return "Evening"

    else:

        return "Other"
```

```python
df["traffic_period"] = df["hour"].apply(traffic_period)
```

```python
period = (

    df.groupby(

        ["vehicle_category","traffic_period"]

    )["total_traffic"]

    .mean()

    .reset_index()

)

period
```

```python
plt.figure(figsize=(12,6))

sns.barplot(

    data=period,

    x="vehicle_category",

    y="total_traffic",

    hue="traffic_period"

)

plt.xticks(rotation=45)

plt.title("Morning vs Evening Traffic")

plt.show()
```

```python
top = (

    df.groupby(

        ["vehicle_category","hour"]

    )["total_traffic"]

    .mean()

    .sort_values(

        ascending=False

    )

    .head(10)

)

top
```

```python
pivot = pd.pivot_table(

    df,

    values="total_traffic",

    index="vehicle_category",

    columns="hour",

    aggfunc="mean"

)

pivot
```

```python
plt.figure(figsize=(14,6))

sns.heatmap(

    pivot,

    cmap="YlOrRd",

    annot=False

)

plt.title("Hourly Traffic Heatmap")

plt.show()
```

```python
summary = (

    peak_hours

    [["vehicle_category","hour","total_traffic"]]

    .rename(

        columns={

            "hour":"Peak Hour",

            "total_traffic":"Average Traffic"

        }

    )

)

summary
```
