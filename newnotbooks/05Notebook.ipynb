```python
PROJECT_DIR = Path.cwd().parent

DATA = PROJECT_DIR/"processed"/"cleaned_dataset.csv"

df = pd.read_csv(DATA)

print(df.shape)

df.head()
```

```python
print("="*60)

print("Vehicle Categories")

print("="*60)

print(df["Vehicle_Category"].unique())
```

```python
hourly_profile = (

    df

    .groupby(

        ["Vehicle_Category","Hour"]

    )["Total_Traffic"]

    .mean()

    .reset_index()

)

hourly_profile = hourly_profile.sort_values(

    ["Vehicle_Category","Hour"]

)

hourly_profile.head()
```

```python
categories = hourly_profile["Vehicle_Category"].unique()

for category in categories:

    temp = hourly_profile[
        hourly_profile["Vehicle_Category"] == category
    ]

    plt.figure(figsize=(10,5))

    plt.plot(

        temp["Hour"],

        temp["Total_Traffic"],

        marker="o"

    )

    plt.xticks(range(24))

    plt.xlim(0,23)

    plt.grid(True)

    plt.title(f"Hourly Traffic Profile - {category}")

    plt.xlabel("Hour")

    plt.ylabel("Average Traffic")

    plt.show()
```

```python
plt.figure(figsize=(14,6))

for category in hourly_profile["Vehicle_Category"].unique():

    temp = hourly_profile[
        hourly_profile["Vehicle_Category"]==category
    ]

    plt.plot(

        temp["Hour"],

        temp["Total_Traffic"],

        marker="o",

        label=category

    )

plt.legend()

plt.xticks(range(24))

plt.xlim(0,23)

plt.grid(True)

plt.xlabel("Hour")

plt.ylabel("Average Traffic")

plt.title("Comparison of Daily Traffic Profiles")

plt.show()
```

```python
peak_hours = (

    hourly_profile

    .loc[

        hourly_profile

        .groupby("Vehicle_Category")

        ["Total_Traffic"]

        .idxmax()

    ]

)

peak_hours = peak_hours.sort_values(

    "Total_Traffic",

    ascending=False

)

peak_hours
```

```python
plt.figure(figsize=(10,5))

plt.bar(

    peak_hours["Vehicle_Category"],

    peak_hours["Total_Traffic"]

)

plt.xticks(rotation=45)

plt.ylabel("Average Traffic")

plt.title("Peak Traffic by Vehicle Category")

plt.show()
```

```python
period = (

    df

    .groupby(

        [

            "Vehicle_Category",

            "Peak_Period"

        ]

    )["Total_Traffic"]

    .mean()

    .reset_index()

)

period
```

```python
pivot = period.pivot(

    index="Vehicle_Category",

    columns="Peak_Period",

    values="Total_Traffic"

)

pivot.plot(

    kind="bar",

    figsize=(12,6)

)

plt.ylabel("Average Traffic")

plt.title("Traffic by Peak Period")

plt.xticks(rotation=45)

plt.show()
```

```python
top10 = (

    df

    .groupby(

        [

            "Vehicle_Category",

            "Hour"

        ]

    )["Total_Traffic"]

    .mean()

    .sort_values(

        ascending=False

    )

    .head(10)

)

top10
```

```python
heatmap = pd.pivot_table(

    df,

    values="Total_Traffic",

    index="Vehicle_Category",

    columns="Hour",

    aggfunc="mean"

)

heatmap
```

```python
plt.figure(figsize=(14,6))

sns.heatmap(

    heatmap,

    cmap="YlOrRd",

    linewidths=.5

)

plt.title("Hourly Traffic Heatmap")

plt.show()
```

```python
summary = (

    df

    .groupby("Vehicle_Category")

    ["Total_Traffic"]

    .agg(

        [

            "mean",

            "max",

            "min",

            "std"

        ]

    )

    .round(2)

)

summary
```

```python
most_busy = (

    df

    .groupby("Vehicle_Category")

    ["Total_Traffic"]

    .mean()

    .sort_values(

        ascending=False

    )

)

most_busy
```

```python
plt.figure(figsize=(12,5))

most_busy.plot(

    kind="bar"

)

plt.ylabel("Average Traffic")

plt.title("Average Traffic by Vehicle Category")

plt.xticks(rotation=45)

plt.show()
```

```python
peak_summary = peak_hours.rename(

    columns={

        "Hour":"Peak Hour",

        "Total_Traffic":"Average Traffic"

    }

)

peak_summary
```

```python
print("="*70)

print("TASK 1 SUMMARY")

print("="*70)

for _,row in peak_summary.iterrows():

    print(

        f"{row['Vehicle_Category']} "

        f"shows the highest average traffic "

        f"around {int(row['Peak Hour'])}:00 "

        f"with approximately "

        f"{round(row['Average Traffic'],2)} vehicles."

    )
```

```python
REPORT_DIR = PROJECT_DIR/"reports"

REPORT_DIR.mkdir(exist_ok=True)

summary.to_csv(

    REPORT_DIR/"vehicle_category_statistics.csv"

)

peak_summary.to_csv(

    REPORT_DIR/"vehicle_peak_hours.csv",

    index=False

)

print("Results Saved Successfully!")
```
