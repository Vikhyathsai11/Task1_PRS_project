```python
import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option("display.max_columns", None)
```

```python
PROJECT_DIR = Path.cwd().parent

DATA = PROJECT_DIR / "processed" / "master1_dataset.csv"

df = pd.read_csv(DATA)

print(df.shape)

df.head()
```

```python
print("="*60)
print("DATASET INFORMATION")
print("="*60)

df.info()
```

```python
missing = pd.DataFrame({

    "Missing Values": df.isnull().sum(),

    "Percentage (%)": round(df.isnull().mean()*100,2)

})

missing.sort_values(
    "Missing Values",
    ascending=False
)
```

```python
#remove duplicates
print("Before :", len(df))

df = df.drop_duplicates()

print("After :", len(df))
```

```python
#missing traffic values
metadata_cols = [

    "Week",

    "Measurement_Site",

    "Latitude",

    "Longitude",

    "Source_File",

    "Vehicle_Category",

    "Time",

    "Total_Traffic"

]

traffic_cols = [

    c

    for c in df.columns

    if c not in metadata_cols

]

df[traffic_cols] = df[traffic_cols].fillna(0)
```

```python
for col in traffic_cols:

    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )
```

```python
df["Total_Traffic"] = df[traffic_cols].sum(axis=1)
```

```python
df["Time"] = pd.to_datetime(
    df["Time"],
    format="%H:%M:%S",
    errors="coerce"
)
```

```python
df["Hour"] = df["Time"].dt.hour

df["Minute"] = df["Time"].dt.minute
```

```python
def peak_period(hour):

    if 6 <= hour < 10:

        return "Morning Peak"

    elif 10 <= hour < 15:

        return "Midday"

    elif 15 <= hour < 19:

        return "Evening Peak"

    else:

        return "Off Peak"


df["Peak_Period"] = df["Hour"].apply(peak_period)
```

```python
df["Time_Interval"] = (

    df["Hour"].astype(str).str.zfill(2)

    + ":"

    + df["Minute"].astype(str).str.zfill(2)

)
```

```python
print("="*60)

print("Rows :", len(df))

print("Columns :", len(df.columns))

print("Vehicle Categories :", df["Vehicle_Category"].nunique())

print("Measurement Sites :", df["Measurement_Site"].nunique())

print("Peak Periods")

print(df["Peak_Period"].value_counts())

print("="*60)
```

```python
df = df.sort_values(

    [

        "Week",

        "Measurement_Site",

        "Vehicle_Category",

        "Time"

    ]

).reset_index(drop=True)
```

```python
OUTPUT = PROJECT_DIR / "processed"

OUTPUT.mkdir(exist_ok=True)

df.to_csv(

    OUTPUT / "cleaned_dataset.csv",

    index=False

)

print("Clean dataset saved successfully!")
```
