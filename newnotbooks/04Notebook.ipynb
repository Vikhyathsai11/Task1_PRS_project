```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

plt.style.use("ggplot")

pd.set_option("display.max_columns", None)
```

```python
PROJECT_DIR = Path.cwd().parent

DATA = PROJECT_DIR / "processed" / "cleaned_dataset.csv"

df = pd.read_csv(DATA)

print(df.shape)

df.head()
```

```python
print("="*70)

print("DATASET INFORMATION")

print("="*70)

df.info()
```

```python
df.describe(include="all")
```

```python
missing = pd.DataFrame({

    "Missing Values": df.isnull().sum(),

    "Percentage (%)": round(df.isnull().mean()*100,2)

})

missing.sort_values(

    "Missing Values",

    ascending=False
)
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows :", duplicates)
```

```python
plt.figure(figsize=(12,5))

df["Vehicle_Category"].value_counts().plot(

    kind="bar"

)

plt.title("Vehicle Category Distribution")

plt.xlabel("Vehicle Category")

plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,5))

df["Measurement_Site"].value_counts().plot(

    kind="bar"

)

plt.title("Measurement Site Distribution")

plt.xlabel("Measurement Site")

plt.ylabel("Records")

plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(

    data=df,

    x="Peak_Period"

)

plt.title("Traffic Records by Peak Period")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(

    df["Total_Traffic"],

    bins=50,

    kde=True

)

plt.title("Distribution of Total Traffic")

plt.show()
```

```python
hourly = (

    df.groupby("Hour")["Total_Traffic"]

    .mean()

)

plt.figure(figsize=(10,5))

plt.plot(

    hourly.index,

    hourly.values,

    marker="o"

)

plt.xticks(range(24))

plt.grid(True)

plt.title("Average Hourly Traffic")

plt.xlabel("Hour")

plt.ylabel("Average Traffic")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(

    data=df,

    x="Vehicle_Category",

    y="Total_Traffic"

)

plt.xticks(rotation=45)

plt.title("Traffic Distribution by Vehicle Category")

plt.show()
```

```python
site = (

    df.groupby("Measurement_Site")["Total_Traffic"]

    .sum()

    .sort_values(

        ascending=False

    )

    .head(10)

)

plt.figure(figsize=(12,6))

site.plot(kind="bar")

plt.ylabel("Total Traffic")

plt.title("Top 10 Busiest Measurement Sites")

plt.show()
```

```python
peak = (

    df.groupby("Peak_Period")["Total_Traffic"]

    .mean()

)

peak
```

```python
plt.figure(figsize=(8,5))

peak.plot(kind="bar")

plt.ylabel("Average Traffic")

plt.title("Average Traffic by Peak Period")

plt.show()
```

```python
traffic_cols = [

    c

    for c in df.columns

    if c not in [

        "Week",

        "Measurement_Site",

        "Latitude",

        "Longitude",

        "Source_File",

        "Vehicle_Category",

        "Time",

        "Peak_Period",

        "Time_Interval"

    ]

]

corr = df[traffic_cols].corr()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(

    corr,

    cmap="coolwarm",

    center=0

)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(

    y=df["Total_Traffic"]

)

plt.title("Outlier Detection")

plt.show()
```
