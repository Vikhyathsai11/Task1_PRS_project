```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

plt.style.use("ggplot")
```

```python
PROJECT_DIR = Path.cwd().parent

df = pd.read_csv(
    PROJECT_DIR / "processed" / "cleaned_dataset.csv"
)

print(df.shape)

df.head()
```

```python
top20 = df.sort_values(
    "Total_Traffic",
    ascending=False
).head(20)

top20
```

```python
top_sites = (

    df

    .groupby("Measurement_Site")

    ["Total_Traffic"]

    .sum()

    .sort_values(

        ascending=False

    )

)

top_sites
```

```python
plt.figure(figsize=(12,5))

top_sites.plot(kind="bar")

plt.ylabel("Total Traffic")

plt.title("Overall Traffic by Measurement Site")

plt.show()
```

```python
vehicles = (

    df

    .groupby("Vehicle_Category")

    ["Total_Traffic"]

    .sum()

    .sort_values(

        ascending=False

    )

)

vehicles
```

```python
plt.figure(figsize=(12,5))

vehicles.plot(kind="bar")

plt.ylabel("Total Traffic")

plt.title("Overall Traffic by Vehicle Category")

plt.show()
```

```python
hourly = (

    df

    .groupby("Hour")

    ["Total_Traffic"]

    .mean()

)

hourly
```

```python
plt.figure(figsize=(12,5))

plt.plot(

    hourly.index,

    hourly.values,

    marker="o"

)

plt.xticks(range(24))

plt.grid(True)

plt.xlabel("Hour")

plt.ylabel("Average Traffic")

plt.title("Overall Hourly Traffic Pattern")

plt.show()
```

```python
peak_hour = hourly.idxmax()

peak_value = hourly.max()

print("Peak Hour :", peak_hour)

print("Average Traffic :", round(peak_value,2))
```

```python
period = (

    df

    .groupby("Peak_Period")

    ["Total_Traffic"]

    .mean()

)

period
```

```python
plt.figure(figsize=(8,5))

period.plot(kind="bar")

plt.ylabel("Average Traffic")

plt.title("Average Traffic by Peak Period")

plt.show()
```

```python
summary = pd.DataFrame({

    "Metric":[

        "Total Records",

        "Survey Weeks",

        "Measurement Sites",

        "Vehicle Categories",

        "Average Traffic",

        "Maximum Traffic",

        "Minimum Traffic"

    ],

    "Value":[

        len(df),

        df["Week"].nunique(),

        df["Measurement_Site"].nunique(),

        df["Vehicle_Category"].nunique(),

        round(df["Total_Traffic"].mean(),2),

        df["Total_Traffic"].max(),

        df["Total_Traffic"].min()

    ]

})

summary
```

```python
REPORT_DIR = PROJECT_DIR/"reports"

REPORT_DIR.mkdir(exist_ok=True)

summary.to_csv(

    REPORT_DIR/"final_summary.csv",

    index=False

)

top_sites.to_csv(

    REPORT_DIR/"overall_site_ranking.csv"

)

vehicles.to_csv(

    REPORT_DIR/"overall_vehicle_ranking.csv"

)

print("Report Saved Successfully")
```
